```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Анализ данных

```python
ds = pd.read_excel('cancer-testicular.xlsx')
ds.head()
```

```python
ds = ds.dropna(axis = 1) 
ds.head()
```

```python
map = ds.iloc[:, 0].tolist()
data = {}
for a,elem in enumerate(map):
    data[elem] = ds.iloc[a, 1:].tolist() 
df = pd.DataFrame(data)
```

```python
# Устанавливаем параметры изображений
import matplotlib as mpl
mpl.rcParams['axes.labelsize'] = 'medium'
mpl.rcParams['figure.labelsize'] = 'large'
mpl.rcParams['xtick.labelsize'] = 'medium'
mpl.rcParams['ytick.labelsize'] = 'medium'
mpl.rcParams['axes.titlesize'] = 'xx-large'
mpl.rcParams['figure.titlesize'] = 'xx-large'
```

```python
import os 
import matplotlib as mpl

os.system('mkdir cancer-testicular-new')

deli = 40
for a in range(int(df.shape[1]/deli)):    
    df[df.columns[deli*a:deli*(a+1)]].hist(figsize=(20, 13.3), bins = 11, layout = (5,8))
    plt.savefig(f'cancer-interferon-new/cancer_testicular_{a}.png',format = 'png')
    plt.show()
df[df.columns[(a+1)*deli:]].hist(figsize=(20, 8), bins = 11, layout = (4,8))
plt.savefig(f'cancer-testicular-new/cancer_testicular_{a+1}.png',format = 'png')
plt.show()    
```

# Убираем гены с явно примитивным распределением

```python
zero_condition = 350
```

```python
drop_columns = []
for i in range(df.shape[1]):
    hist = np.histogram(list(df.iloc[:,i]), bins=11, range=None, density=None, weights=None)[0]
    if(hist[0] > zero_condition):
        print(f'{i} ', end = '')
        drop_columns.append(df.iloc[:,i].name)
```

```python
drop_columns
```

```python
df_drop = df.drop(columns = drop_columns)
deli = 40
os.system('mkdir cancer-testicular-hist-without-primitive')
for a in range(int(df_drop.shape[1]/deli)):    
    df_drop[df_drop.columns[deli*a:deli*(a+1)]].hist(figsize=(20, 13.3), bins = 11, layout = (5,8))
    plt.savefig(f'cancer-testicular-hist-without-primitive/cancer_testicular_{a}.png',format = 'png')
    plt.show()
df_drop[df_drop.columns[(a+1)*deli:]].hist(figsize=(20, 8), bins = 11, layout = (3,8))
plt.savefig(f'cancer-testicular-hist-without-primitive/cancer_testicular_{a+1}.png',format = 'png')
plt.show()    
```
